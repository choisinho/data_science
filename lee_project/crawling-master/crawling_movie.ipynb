```python
import numpy as np
import pandas as pd
import requests as rq #url을 가져오기 위한 파이썬 내장 패키지

from bs4 import BeautifulSoup
```

```python
url = rq.get('https://movie.naver.com/movie/bi/mi/review.naver?code=192608' )
#영화 백두산 평점 리스트 사이트 주소
print(url) #Response[200]이 출력되면 성공적으로 사이트를 가져옴
```

```python
soup = BeautifulSoup(url.text, 'html.parser')
```

```python
print(soup)
```

```python
import re
score_list = []
def croll_score() :
    for score in soup.find_all('div','list_netizen_score'):
        score_list.append(int(re.findall('\d+', str(score.get_text))[2]))
```

```python
croll_score()
print(score_list)
```

```python
review_list = []
for review in soup.find_all('td','title'):
    review_list.append(review.get_text().split('\n')[5])
```
