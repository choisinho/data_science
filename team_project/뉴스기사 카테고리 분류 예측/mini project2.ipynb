# 각 컬럼 별 기사 1000개씩 크롤링

```python
# 부동산 기사 크롤링
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import subprocess

subprocess.Popen(r'C:\Program Files\Google\Chrome\Application\chrome.exe --remote-debugging-port=9222 --user-data-dir="C:\chrometemp"') # 디버거 크롬 구동

option = Options()
option.add_experimental_option("debuggerAddress", "127.0.0.1:9222")

driver = webdriver.Chrome(f'chromedriver.exe', options=option)
driver.implicitly_wait(10)

news_land = {'title' : [], 'body' : []}
start = 2660000 #최근 부동산 기사 시작
counter = 0

while counter < 1000:
    try:
        driver.get("https://land.naver.com/news/newsRead.naver?type=headline&bss_ymd=&prsco_id=119&arti_id=000{:d}".format(start))

        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')

        title = soup.find('div', class_='article_header')
        title_text = title.get_text(strip=True)

        body = soup.find('div', class_='article_body size4')
        body_text = body.get_text(strip=True)
        
        news_land['body'].append(body_text)
        news_land['title'].append(title_text)
        counter += 1
    except:
        pass
    
    start += 1
    print("counter : ", counter, "| start : ", start)
    time.sleep(1)
```

```python
# 연예 기사 크롤링

import time
from selenium import webdriver
from bs4 import BeautifulSoup

subprocess.Popen(r'C:\Program Files\Google\Chrome\Application\chrome.exe --remote-debugging-port=9222 --user-data-dir="C:\chrometemp"') # 디버거 크롬 구동

option = Options()
option.add_experimental_option("debuggerAddress", "127.0.0.1:9222")

driver = webdriver.Chrome(f'chromedriver.exe', options=option)
driver.implicitly_wait(10)

news_entertain = {'title' : [], 'body' : []}
start = 4706847
counter = 0

while counter < 1000:
    driver.get("https://entertain.naver.com/now/read?oid=109&aid=000{:d}".format(start))

    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    title = soup.find('h2', class_='end_tit')
    title_text = title.get_text(strip=True)
    
    if '[사진]' in title_text: #사진만 있는 기사는 PASS
        pass
    else:
        body = soup.find('div', class_='end_body_wrp')
        try:
            body_text = body.get_text(strip=True)
        except:
            body_text = ''
        news_entertain['body'].append(body_text)
        news_entertain['title'].append(title_text)
        counter += 1
        
    start += 1
    print("counter : ", counter, "| start : ", start)
    time.sleep(1)
```

```python
import pandas as pd

land = pd.DataFrame(news_land)
land.to_csv('부동산.csv', index=False)
enter = pd.DataFrame(news_entertain)
enter.to_csv('연예.csv', index=False)
```

```python
# 부동산 : 0, 연예 : 1
land['label'] = 0
enter['label'] = 1
```

```python
join = land.append([enter])
join.reset_index(inplace=True)
join.drop('index', axis=1, inplace=True)

def place2(a):
    return a.replace('이미지 원본보기', '')

join ['body'] = join['body'].map(place2)
```

# 기사 내용으로 예측

```python
from sklearn.model_selection import train_test_split

train_x, test_x, train_y, test_y = train_test_split(join['body'], join['label'], test_size=0.5)
```

```python
from keras.preprocessing.text import Tokenizer

token = Tokenizer(num_words=10000)
token.fit_on_texts(train_x)
train_x = token.texts_to_matrix(train_x)
test_x = token.texts_to_matrix(test_x)
```

```python
from keras.utils import to_categorical

train_y = to_categorical(train_y)
test_y = to_categorical(test_y)
```

```python
len(train_x)
```

```python
from keras import models
from keras import layers

model = models.Sequential()
model.add(layers.Dense(64, activation='relu', input_shape=(10000,)))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(2, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(train_x, train_y, epochs=10, batch_size=512, validation_split=0.2)
```

```python
model.evaluate(test_x, test_y)
```

# 기사 제목으로 예측

```python
from sklearn.model_selection import train_test_split

train_x, test_x, train_y, test_y = train_test_split(join['title'], join['label'], test_size=0.5)
```

```python
from keras.preprocessing.text import Tokenizer

token = Tokenizer(num_words=10000)
token.fit_on_texts(train_x)
train_x = token.texts_to_matrix(train_x)
test_x = token.texts_to_matrix(test_x)
```

```python
from keras.utils import to_categorical

train_y = to_categorical(train_y)
test_y = to_categorical(test_y)
```

```python
from keras import models
from keras import layers

model = models.Sequential()
model.add(layers.Dense(64, activation='relu', input_shape=(10000,)))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(2, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(train_x, train_y, epochs=20, batch_size=512, validation_split=0.2)
```

```python
model.evaluate(test_x, test_y)
```
